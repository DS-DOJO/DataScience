# Numpy vs List

Import library

```python
import matplotlib.pyplot as plt
import numpy as np 
import time
```

Speed comparision

```python
def time_list_operation(size, operation):

    create_list_1 = range(size)
    create_list_2 = range(size)
    
    # capturing time before the addition of number to Python lists 
    initialTime_list = time.time()

    # start operation
    if operation == 'div':
        new_list = [create_list_1[i] / (create_list_2[i]+1) for i in range(len(create_list_1))]
    elif operation == 'sub':
        new_list = [create_list_1[i] - create_list_2[i] for i in range(len(create_list_1))]
    elif operation == 'mul':
        new_list = [create_list_1[i] * create_list_2[i] for i in range(len(create_list_1))]
    else : 
        new_list = [create_list_1[i] + create_list_2[i] for i in range(len(create_list_1))]


    # capturing time afetr the addition of number to Python lists 
    finalTime_list = round(time.time() - initialTime_list,10)

    #print(finalTime_list)
    return finalTime_list
```
